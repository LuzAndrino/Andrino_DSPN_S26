### Writing a function - R

```r
fahrenheit_to_celsius <- function(temp_F){
    tempC <-temp_F -32 *5/9
    return(tempC)
}
```

```r
fahrenheit_to_celsius(32)
```

```r
fahrenheit_to_celsius(102)
```

```r
input1 <- 20 
my_sum <- function(input1, input2= 10){
    output <- input1 + input2
    return(output)
}
```

```r
my_sum(2)
```

```r
my_sum(3,4)

my_sum(input1 = 3)
```
