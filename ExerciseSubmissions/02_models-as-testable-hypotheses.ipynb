## 1. Summary statistics (4 pts)

Write a function that takes a vector of numbers `x` and returns the length, mean, and standard deviation of `x` as a new vector.

In keeping with our best practices, give the function a short but descriptive name, and use snake case if it involves multiple words.

Hint: Vectors are defined in R using the `c()` command.

```r
#Write your function here
x <-4, 2, 5, ,3 
```

```r
x  <- c(2,3,4,5)
```

```r

summary_stats <- function(x) {
    length_x  <- length(x)
    mean_x  <- mean(x)
    sd_x  <- sd(x)
    return(c(length_x, mean_x, sd_x))
}
print(paste("Summary statistics for x:",summary_stats(x)))
```

```r
#New and improved summary statistics function 
```

```r
summary_stats  <- function (x) {
    length_x  <- length(x[!is.na(x)]) #ignores NA values uses only true values
    mean_x  <- mean(x,na.rm = TRUE) #removes NA's 
    sd_x  <- sd(x,na.rm = TRUE)
    return(c(length_x, mean_x, sd_x))
}
summary_stats(v1)
print(paste("Summary statistics for v1:",summary_stats(v1)))
```

## 2. T-test function (4 pts)

The formula for a t-test is:

```r
x <- c(2, 3, 4, 5)
ttest_fun  <- function(x, mu) {#have 2 arguments as told by the Hint 
    summary_stats  <- summary_stats(x) #length, mean, std dev
    n  <- summary_stats [1] #length
    m <- summary_stats [2] #mean
    s  <- summary_stats [3] #std_dev
    t_stat <- (m - mu)/(s/sqrt(n))
}
```

```r
attempt_1  <- ttest_fun(x, mu = 3)
print(paste("T-test 1 =", attempt_1))
```

$$ \frac{m- \mu}{ \frac{s}{\sqrt{n}}} $$

Where m is the sample mean, $\mu$ (mu) is the population mean, s is the standard deviation, and n is the sample size.
